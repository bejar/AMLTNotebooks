# Time Series and Frequent Sequences

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
%matplotlib notebook
```

```python
datapath = '../Data/'

data = np.loadtxt(datapath+'tseries.csv')
fig = plt.figure(figsize=(12,6))
plt.plot(data)
```

```python
import amltlearn.time_series.discretization.SAX as SAX
saxtrans = SAX(window_length=data.shape[0], step=1, word_length=1000, voc_size=15)
datatrans = saxtrans.transform(data)
```

```python
voc = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
off = int(min(datatrans[0]))
datastr = ""
for v in datatrans[0]:
    datastr += voc[int(v)-off]
datastr
```

```python
from rstr_max import *
import operator

rstr = Rstr_max()
rstr.add_str(datastr)

r = rstr.go()
lstrings = []
sup =  25

for (offset_end, nb), (l, start_plage) in r.items():
    ss = rstr.global_suffix[offset_end - l:offset_end]
    id_chaine = rstr.idxString[offset_end - 1]
    s = rstr.array_str[id_chaine]
    if nb > sup and len(ss.encode('utf-8')) > 1:
        lstrings.append((ss.encode('utf-8'), nb))
   
lstrings = sorted(lstrings, key=operator.itemgetter(0), reverse=False)
for s in lstrings:
    print(s)
```
